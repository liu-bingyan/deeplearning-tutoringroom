```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constant beta value
beta = 598

# Define the function T(t) = integral of beta(s) from 0 to t
# Since beta is constant, T(t) = beta * t
def T(t):
    return beta * t

# Define alpha_t = e^(-1/2 * T(1-t))
def alpha_t(t):
    return np.exp(-0.5 * T(1-t))

# Define sigma_t = 1 - alpha_t^2
def sigma_t(t):
    return 1 - alpha_t(t)**2

# Create t values from 0 to 1
t = np.linspace(0, 1, 1000)

# Calculate alpha_t and sigma_t values
alpha_values = alpha_t(t)
sigma_values = sigma_t(t)

# Create the plots
plt.figure(figsize=(12, 5))

# Plot for alpha_t
plt.subplot(1, 2, 1)
plt.plot(t, alpha_values, 'b-', linewidth=2)
plt.title(r'$\alpha_t = e^{-\frac{1}{2} T(1-t)}$')
plt.xlabel('t')
plt.ylabel(r'$\alpha_t$')
plt.grid(True)

# Plot for sigma_t
plt.subplot(1, 2, 2)
plt.plot(t, sigma_values, 'r-', linewidth=2)
plt.title(r'$\sigma_t = 1 - \alpha_t^2$')
plt.xlabel('t')
plt.ylabel(r'$\sigma_t$')
plt.grid(True)

plt.tight_layout()
plt.savefig('alpha_sigma_plots.png', dpi=300)
plt.show()

# Print some values for verification
print(f"At t=0: alpha_t = {alpha_t(0):.6f}, sigma_t = {sigma_t(0):.6f}")
print(f"At t=0.5: alpha_t = {alpha_t(0.5):.6f}, sigma_t = {sigma_t(0.5):.6f}")
print(f"At t=1: alpha_t = {alpha_t(1):.6f}, sigma_t = {sigma_t(1):.6f}")
```
